Automatic File Sorter in File Explorer

```python
import os, shutil
```

```python
path = r"C:/dysk/CV â€” kopia/Naukaa/Python/Data/"
```

```python
file_name = os.listdir(path)
```

```python
folder_names = ['csv_files', 'text_files', 'image_files']   

for loop in range(0,3):
    if not os.path.exists(path + folder_names[loop]):
        print(path + folder_names[loop] + " created")
        os.makedirs(path + folder_names[loop])

for file in file_name:
    if file.endswith('.csv') and not file.startswith('csv_files'):
        shutil.move(path + file, path + 'csv_files/' + file)
        print(f"{file} moved to csv_files folder")
    elif file.endswith('.txt') and not file.startswith('text_files'):
        shutil.move(path + file, path + 'text_files/' + file)
        print(f"{file} moved to text_files folder")
    elif file.endswith('.png') or file.endswith('.jpg') or file.endswith('.jpeg')  and not file.startswith('image_files'):
        shutil.move(path + file, path + 'image_files/' + file)
        print(f"{file} moved to image_files folder")
    
```
