```python
def number_args(*args):
    print(args[0]*args[1])
```

```python
number_args(1,2,3,4,5)
```

Jezeli chcemy Tworzyc slownik jako klucz

```python
def number_kwarg(**number):
    print('My number is: ' + number['integer'] + 'My other number:' + number['integer2'])
```

```python
number_kwarg(integer = '2309', integer2 = '349')
```

1) Co robi **number w definicji funkcji?
def number_kwarg(**number):


**number oznacza: „zbierz wszystkie argumenty nazwane przekazane do funkcji i włóż je do słownika”.

Czyli jeśli wywołasz:

number_kwarg(integer='2309', integer2='349')


to wewnątrz funkcji Python zrobi z tego słownik:

number = {
  'integer': '2309',
  'integer2': '349'
}
