Set data type

```python
#set data type
#pokazujee tylko unikalne wartosci
my_set = {2, 4, 6, 8, 10}
```

```python
type(my_set)
```

```python
print(my_set)
```

```python
my_sets= {2, 4, 6, 8, 10, 2, 4}
print(my_sets)
```

```python
test_set = {1, 2, 3, 4, 5}
#pokazuje wszystkie elementy ktore sa w obu setach
print(my_sets | test_set)  # union
```

```python
#pokazuje tylko elementy ktore sa w obu setach
print(my_sets & test_set)  # intersection
```

```python
print(my_sets - test_set)  # difference
```

```python
print(my_sets ^ test_set)  # tych elementow ktorych nie ma w obu setach
```
