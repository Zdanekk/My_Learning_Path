BMI Calculator

```python
weight = float(input("Enter your weight in kilograms (kg): "))
print("Your height: ", weight)

height = float(input("Enter your height in centimeters (cm): "))
print("Your height: ", height)

# basic input validation (example ranges)
if weight < 20 or weight > 350:
    print("Weight is out of range (20–350 kg).")
elif height < 80 or height > 250:
    print("Height is out of range (80–250 cm).")
else:
    bmi = weight / ((height / 100) ** 2)
    print("Your BMI is:", round(bmi, 2))

    if bmi < 18.5:
        print("Underweight")
    elif bmi < 25:
        print("Normal weight")
    elif bmi < 30:
        print("Overweight")
    elif bmi < 35:
        print("Obesity (Class I)")
    elif bmi < 40:
        print("Obesity (Class II)")
    else:
        print("Obesity (Class III)")
```
