For loop

```python
integers = [1, 2, 3, 4, 5]
```

```python
for number in integers:
    print(number)
```

```python
for jellly in integers:
    print(jellly + jellly)
```

```python
ice_cream_dict={ 'name': 'Ice Cream', 'ingredients': ['milk', 'sugar', 'flavoring'], 'price': 5.99 }
```

```python
for cream in ice_cream_dict.values():
    print(cream)
```

```python
for key, value in ice_cream_dict.items():
    print(key,'->', value)
```

Nested For Loops

```python
flavors = ['vanilla', 'chocolate', 'strawberry']
toppings = ['sprinkles', 'chocolate chips', 'caramel']
for flavor in flavors:
    for topping in toppings:
        print(f"{flavor} ice cream with {topping}")
```

While loops

```python
number = 0
while number < 5:
    print(number)
    number += 1
```

```python
number = 0
while number < 5:
    print(number)
    if number == 3:
        break
    number += 1
```

```python
number = 0
while number < 5:
    print(number)
    if number == 5:
        break
    number += 1
else:
    print('No longer less than 5')
```

```python
number = 0
while number < 5:
    number += 1
    if number == 3:
        continue
    print(number)
else:
    print('No longer less than 5')
```
